1. Crie uma nova coluna chamada: “house_age”. 
    * Se o valor da coluna “date” for maior que 2014-01-01 => ‘new_house’
    * Se o valor da coluna “date” for menor que 2014-01-01 => ‘old_house’.

```python
import numpy as np
import pandas as pd

data = pd.read_csv('/kc_house_data.csv')
print( data.columns)
```

```python
# Criando a coluna  “house_age” no dataset

data['fabiano'] = "house_age"
print( data.columns)
```

```python
data.head(7)
```

```python
data.loc[data['date'] > '2014-01-01', 'fabiano'] = 'new_house' 
data.loc[data['date'] < '2014-01-01', 'fabiano'] = 'old_house' 
```

```python
data.head(7)
```

2. Crie uma nova coluna chamada “dormitory_type”.
    * Se o valor da coluna “bedrooms” for igual à 1 => ‘studio’
    * Se o valor da coluna “bedrooms” for igual a 2 => ‘apartament’
    * Se o valor da coluna “bedrooms” for maior que 2 => ‘house’.

```python
# Criando a coluna  “dormitory_type” no dataset
data['dormitory'] = "dormitory_type"
print( data.columns)
data.head(7)
```

```python
data.loc[data['bedrooms'] == 1, 'dormitory'] = 'studio' 
data.loc[data['bedrooms'] == 2, 'dormitory'] = 'apartament' 
data.loc[data['bedrooms'] > 2,'dormitory'] = 'house'

data.head(7)
```
