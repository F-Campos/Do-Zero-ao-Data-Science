```python
import pandas as pd
```

```python
data = pd.read_csv('/kc_house_data.csv')
```

```python
data.head()
```

```python
# Exemplo: Quantos imóveos existem por numero de quartos?

data['bedrooms'] == 0  # gera uma sequência de index
```

```python
data[data['bedrooms'] == 0]  #filtro todas as linhas onde 'bedrooms' é igual 0
```

```python
data[data['bedrooms'] == 0].shape # Assim temos 13 imíveis 'com 0 quartos'
```

```python
print( data[data['bedrooms'] == 0].shape )
print( data[data['bedrooms'] == 1].shape )
print( data[data['bedrooms'] == 2].shape )
print( data[data['bedrooms'] == 3].shape )
print( data[data['bedrooms'] == 4].shape )
```

```python
df_grouped = data[['id', 'bedrooms']].groupby( 'bedrooms' ) 
```

```python
for bedrooms, frame in df_grouped:  #frame é o numero de linhas do meu dataset que form selecionadas a partir do meu filtro
        print( 'numero de quartos: {}'.format( bedrooms ))
        print( frame.shape, end= '\n\n' )
```
